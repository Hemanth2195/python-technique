## printing all permutaions of a given array with non-repeating elements.

problem: https://www.interviewbit.com/problems/permutations/
permutaions of a list using recursion

how recursion goes and base case watch : https://youtu.be/KukNnoN-SoY

i did it by my self the code. 

```python
class Solution:
    # @param A : list of integers
    # @return a list of list of integers
    def permute(self, A):
        start=A
        end=[]
        stack=[[start,end]]
        ans=[]
        def perm(stack):
            p=stack.pop()
            s,e=p[0],p[1]
            if s==[]:
                ans.append(e)
                return 
            for i in range(len(s)):
                s1=s.copy()
                e1=e.copy()
                s1.remove(s[i])
                e1.append(s[i])
                stack.append([s1,e1])
                perm(stack) 
        perm(stack)
        return ans
```

```python
c1=Solution()
c1.permute([1,2,3])
```

## proper parenthesis possible with given no.of parenthesis.

problem: https://www.interviewbit.com/problems/generate-all-parentheses-ii/

```python
class Solution:
    # @param A : integer
    # @return a list of strings
    def generateParenthesis(self, A):
        if A==1:
            return ['()']
        ans=[]
        start=['.']*2*(A-1)
        start=['(']+start+[')']
        stack=[[start,1]] 
        while stack:
            c=stack.pop(0)
            p=c[1]
            c1=c[0].copy()
            c2=c[0].copy()
            c1[p]='('
            c2[p]=')'
            
            C1=c1.count('(')
            D1=c1.count(')')
            if C1==A:
                c1=''.join(c1)
                c1=c1.replace('.',')') 
                ans.append(c1) 
            else:
                stack.append([c1,p+1]) 
            C2=c2.count('(')
            D2=c2.count(')')
            if D2-C2>1: 
                pass
            else:
                stack.append([c2,p+1]) 
        ans.sort()
        return ans 
                
```
